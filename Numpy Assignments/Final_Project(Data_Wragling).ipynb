# 3. Low Level Data Understanding 

# a. Find count of unique values in location column.

```python
print(df['location'].nunique())
```

# b. Find which continent has maximum frequency using values counts.

```python

print(f"\ncontinent having maximum frequency is {df['continent'].value_counts().idxmax()}")

```

# c. Find maximum & mean value in 'total_cases'.

```python
max_value = df['total_cases'].max()
mean_value = df['total_cases'].mean()
print('Maximum value of total_cases:', max_value)
print('\n')
print('Mean value of total_cases:', mean_value)
```

# d. Find 25%,50% & 75% quartile value in 'total_deaths'.

```python
df['total_deaths'].quantile([.25,.5,.75])
print(f"25% of total_death_values : {df['total_deaths'].quantile(q=.25)}")
print(f"50% of total_death_values : {df['total_deaths'].quantile(q=.5)}")
print(f"75% of total_death_values : {df['total_deaths'].quantile(q=.75)}")
```

# e. Find which continent has maximum 'human_development_index'

```python
max_1 = df.groupby('continent')['human_development_index'].max().idxmax()
print(max_1)
```

# f. Find which continent has minimum 'gdp_per_capita'.

```python
max_2 = df.groupby('continent')['gdp_per_capita'].min().idxmin()
print(max_2)

print(df.groupby('continent')['gdp_per_capita'].min().sort_values().head(1))
```

# 9. Data Visualization :

# a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
sns.set()
sns.distplot(x=df['gdp_per_capita'],label="gdp_Per_Capita",color="blue", bins=25)
plt.title('Histogram of GDP_Per_Capita')
plt.legend()
```

# b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

```python
plt.figure(figsize=(16,9))
sns.set()
sns.scatterplot(x=df['total_cases'], y=df['gdp_per_capita'],s=200, edgecolor = 'y', linewidth = 10)
plt.xlabel('Total Cases',)
plt.ylabel('GDP_Per_Capita')
plt.title('Total Cases Upon GDP_Per_Capita')
plt.show()
```
